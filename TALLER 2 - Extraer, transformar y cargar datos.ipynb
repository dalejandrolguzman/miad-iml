- `spark_ordenes_df`: las columnas que describen las órdenes de compra. Estas columnas tienen nombres de la forma `"Order_<campo>"`. <br><br>

    - La columna `"Customer_Code"` define individualmente a cada cliente.<br><br>
    - De cada cliente se encarga un único agente. Con el fin de respetar la relación de negocios entre los agentes y sus respectivos clientes, debemos incluir en este `DataFrame` la columna `Agent_Code`.<br><br>

- `spark_agentes_df`: las columnas que describen a los agentes. Estas columnas tienen nombres de la forma `"Agent_<campo>"`. <br><br>

Asegúrate de ordenar las tablas por la columna que define los registros individualmente.

Ayuda: puedes utilizar el método `DataFrame.colRegex`.

```python
from pyspark.sql.functions import col

def desagregar_df(df):
    # Extraer columnas de ordenes
    ordenes_cols = [c for c in df.columns if c.startswith("Order_") or c in ["Customer_Code", "Agent_Code"]]
    spark_ordenes_df = df.select(ordenes_cols).distinct().orderBy("Order_Number")

    # Extraer columnas de clientes
    clientes_cols = [c for c in df.columns if c.startswith("Customer_") or c == "Agent_Code"]
    spark_clientes_df = df.select(clientes_cols).distinct().orderBy("Customer_Code")

    # Extraer columnas de agentes
    agentes_cols = [c for c in df.columns if c.startswith("Agent_")]
    spark_agentes_df = df.select(agentes_cols).distinct().orderBy("Agent_Code")

    return spark_ordenes_df, spark_clientes_df, spark_agentes_df

spark_ordenes_df, spark_clientes_df, spark_agentes_df = desagregar_df(spark_ventas_df)
```

```python
# prompt: Soluciona ejercicio 3

def crear_varios_db(lista_nombres):
    # Crea una lista para almacenar los nombres de las bases de datos creadas
    nombres_creados = []

    # Itera sobre la lista de nombres de bases de datos
    for nombre in lista_nombres:
        try:
          # Intenta crear la base de datos
          spark.sql(f"CREATE DATABASE IF NOT EXISTS {nombre}")
          nombres_creados.append(nombre)
        except Exception as e:
          print(f"Error al crear la base de datos {nombre}: {e}")
          pass

    # Crea un DataFrame de PySpark con los nombres de las bases de datos creadas
    spark_bases_de_datos_df = spark.createDataFrame(nombres_creados, "string").toDF("namespace")

    return spark_bases_de_datos_df

spark_bases_de_datos_df = crear_varios_db(nombres_a_crear)
spark_bases_de_datos_df.show()
```
